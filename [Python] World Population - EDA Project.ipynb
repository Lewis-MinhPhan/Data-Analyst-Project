### Average world population grouped by continents

```python
# Average world population grouped by continents

df.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```

```python
# Filter all countries in 'Oceania' continent

df[df['Continent'].str.contains('Oceania')].sort_values(by = '2022 Population',ascending = False)
```

### Chart for world population

```python
# Sorting column to display population from 1980 to 2022

df_continents = df.groupby('Continent')[list(reversed(df.columns[5:12]))].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)

df_continents

# Manually sorting columns
# df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
```

```python
df_continents_chart = df_continents.transpose()
```

```python
df_continents_chart.plot()
```
